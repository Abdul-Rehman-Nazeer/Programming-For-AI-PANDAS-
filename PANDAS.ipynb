```python
# q7. Write a Pandas program to import excel data (employee.xlsx ) into a Pandas dataframe and
# find a list of employees of a specified year.

import pandas as pd

df = pd.read_excel("employee.xlsx")

year =int(input("Enter the year you want to search employee"))

filtered_df = df[df['hire_date'].dt.year == year]

if not filtered_df.empty:
    print(f"Employees who joined in {year}:\n")
    print(filtered_df)
else:
    print(f"No employees found for the year {year}.")
```

```python
# q8. You work for a small e-commerce company that sells products online. You have two CSV
# files: one containing product information and another containing order data. Your goal is to
# load and analyze this data using Pandas to answer specific questions. (use csv format to
# create and load data).
# 1) Product Information (products.csv)
# Columns: ProductID, ProductName, Category, Price
# 2) Order Data (orders.csv)
# Columns: Order ID, ProductID, Quantity, Order Date (MM-DD-YYYY)

import pandas as pd
products_df = pd.read_csv('Products.csv')
orders_df = pd.read_csv('orders.csv')
```

```python
merged_df = pd.merge(orders_df, products_df, on='Product ID')
merged_df['TotalRevenue'] = merged_df['Price'] * merged_df['Quantity']
total_revenue = merged_df['TotalRevenue'].sum()
print("Total revenue:", total_revenue)
```

```python
#d
product_sales = merged_df.groupby('Product Name')['Quantity'].sum().sort_values(ascending=False)

top_5_best_selling = product_sales.head(5)
top_5_low_selling = product_sales.tail(5)

print('\nTop 5 best-selling products\n')
print(top_5_best_selling)

print('\nTop 5 low-selling product\n')
print(top_5_low_selling)
```

```python
#e
top_5_categories = merged_df[merged_df['Product Name'].isin(top_5_best_selling.index)]['Category']
most_common_category = top_5_categories.mode().iloc[0]

print('\nMost common category among top 5 best-selling products\n', most_common_category)
```

```python
#f
average_price_per_category = products_df.groupby('Category')['Price'].mean()
print('\nAverage price of products in each category\n')
print(average_price_per_category)
```

```python
#g
highest_revenue_order = merged_df.loc[merged_df['TotalRevenue'].idxmax()]
highest_revenue_order_date = pd.to_datetime(highest_revenue_order['Date'])
highest_revenue_day = highest_revenue_order_date.day
highest_revenue_month = highest_revenue_order_date.month
highest_revenue_year = highest_revenue_order_date.year

print('\nDay with highest total revenue ', highest_revenue_day)
print('Month with highest total revenue ', highest_revenue_month)
print('Year with highest total revenue ', highest_revenue_year)
```

```python
#h
import pandas as pd

merged_df['Date'] = merged_df['Date'].str.strip()

merged_df['Date'] = pd.to_datetime(merged_df['Date'], errors='coerce')

merged_df = merged_df.dropna(subset=['Date'])

merged_df['Month'] = merged_df['Date'].dt.month

revenue_per_month = merged_df.groupby('Month')['TotalRevenue'].sum().reset_index()

revenue_per_month.columns = ['Month', 'TotalRevenue']

print('\nTotal revenue for each month\n')
print(revenue_per_month)

highest_revenue_month = revenue_per_month.loc[revenue_per_month['TotalRevenue'].idxmax()]

print('Month with the highest total revenue:')
print(highest_revenue_month)
```

```python
import pandas as pd
import numpy as np

products_df = pd.read_csv("Products.csv")
orders_df = pd.read_csv("Orders.csv")

products_df = products_df.dropna()
orders_df = orders_df.dropna()

def clean_invalid_characters(column_data):
    column_data = column_data.str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
    return column_data

products_df['Product Name'] = clean_invalid_characters(products_df['Product Name'])

print("Cleaned Products Data:")
print(products_df.head())

print("Cleaned Orders Data:")
print(orders_df.head())

products_df.to_csv("CleanedProducts.csv", index=False)
orders_df.to_csv("CleanedOrders.csv", index=False)
```
